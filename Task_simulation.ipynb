## Задача 1. Пункт б)

```python
random.seed(0)

'-------------------'
people_in_chosen_bus = []
for _ in range(10**4):
    people_in_bus = []
    time = 8*60
    while time < 20*60:
        gap = random.choice([5,10])
        if gap == 5:
            people_in_bus.append(5)
        else:
            people_in_bus.append(10)
        time += gap
    people_in_chosen_bus.append(np.random.choice(people_in_bus))
    
plt.title('Распределение количества пассажиров у Тимона')
plt.hist(people_in_chosen_bus, bins = range(0,20))
plt.xlabel('Количество пассажиров в выбранном автобусе')
plt.ylabel('Количество раз встретилось')
plt.show()
print(f'Матожидание примерно равно {np.mean(people_in_chosen_bus)}')
print(f'Дисперсия примерно равна {np.var(people_in_chosen_bus)}')
```

## Задача 2. Пункт а)

```python
random.seed(0)
track = np.arange(1, 10**3 + 1, dtype = int)

def atos():
    balances = np.empty(0, dtype = int)
    balance_Atos = 1000
    i = 0
    while i < 10**3:
        i +=1
        number = np.random.randint(1,100)
        if number <= 49: 
            balance_Atos += 1
        else:
            balance_Atos -=1
        balances = np.append(balances, balance_Atos)
    return balances
all_matrix = np.empty((10**4, 1000))

# симулируем походы в казино
plt.figure(figsize = (20,10))
for l in range(10 ** 4):
    y = atos()
    all_matrix[l, :] = y
    
# выбираем 10 рандомных походов
matrix = np.empty((10, 1000))
random_trips = np.random.randint(10**4, size = 10)
for i in range(10):
    y = all_matrix[random_trips[i]]
    plt.plot(track, y ,linewidth = 1, markersize = 3, label = f'поход {i+1}', color = 'k')

# дорисовываем график
plt.title('Исход казино А')
plt.legend()
plt.xlabel('Номер похода', fontsize = 15)
plt.ylabel('Баланс Атоса', fontsize = 15)
mean = np.mean(all_matrix, axis=0)
plt.plot(track, mean, linewidth = 5, label='среднее значение', color = 'royalblue')
plt.show()
```

#### Действительно, по вероятностям казино проигрышное. E(x) = 1000 + 1 * 1000(0,49-0,51) = 980. Примерно сходится

# Задача 3 (Парадокс инспектора)

## Задача 3. Пункт а)

```python
from scipy.stats import norm
random.seed(0)
ns = np.arange(80, 165, 5)
proshli = np.arange(0)
for n in ns:
    matrix = np.zeros((3, 10000))
    matrix[0] = norm(60, 10).rvs(10000)
    matrix[1] = norm(60,10).rvs(10000)
    matrix[2] = matrix[0] + matrix[1]
    summa = (matrix[2] > n).sum()
    proshli = np.append(proshli, summa) 

# дорисовываем график
plt.figure(figsize = (20, 10))
plt.title('Число прошедших в УШЭ в зависимости от n', fontsize = 25)
plt.bar(ns, proshli, width = 3)
plt.xlabel('n', fontsize = 15)
plt.ylabel('Количество прошедших в УШЭ', fontsize = 15)
plt.show()
```

#### Больше баллы --> меньше шанс попасть (ﾉ´ з `)ノ
---

```python
ns = np.arange(80, 165, 5)
correlations = np.arange(0)
from scipy import stats as sts

for n in ns:
    matrix = np.zeros((3, 10000))
    matrix[0] = sts.norm(60, 10).rvs(10000)
    matrix[1] = sts.norm(60,10).rvs(10000)
    matrix[2] = matrix[0] + matrix[1]
    new_matrix = matrix[:, matrix[2, :] > n]
    сorrelation = np.corrcoef(new_matrix[0, :], new_matrix[1, :])[0,1]
    correlations = np.append(correlations, сorrelation) 
    
# дорисовываем график
plt.figure(figsize = (20, 10))
plt.title('Сorr между баллами по предметам у прошедших в УШЭ в зависимости от n', fontsize = 25)
plt.plot(ns, correlations, linewidth = 7)
plt.xlabel('n', fontsize = 15)
plt.ylabel('Corr', fontsize = 15)
plt.show()
```

#### 

#### В список прошедших в УШЭ мы часто выбираем тех, кто, например, набрал 80 и 10 баллов ( то есть тех, кто набрал много баллов по одному предмету), поэтому их корреляция часто отрицательна
--- 

```python
ns = np.arange(80, 165, 5)
correlations = np.arange(0)

for n in ns:
    matrix = np.zeros((3, 10000))
    matrix[0] = norm(60, 10).rvs(10000)
    matrix[1] = norm(60,10).rvs(10000)
    matrix[2] = matrix[0] + matrix[1]
    new_matrix = matrix[:, matrix[2, :] <= n]
    сorrelation = np.corrcoef(new_matrix[0, :], new_matrix[1, :])[0,1]
    correlations = np.append(correlations, сorrelation) 
    
# дорисовываем график
plt.figure(figsize = (20, 10))
plt.title('Сorr между баллами по предметам у непрошедших в УШЭ в зависимости от n', fontsize = 25)
plt.plot(ns, correlations, linewidth = 7)
plt.xlabel('n', fontsize = 15)
plt.ylabel('Corr', fontsize = 15)
plt.show()
```

#### В список не прошедших в УШЭ реже попадают те, кто выделился по одному предмету и прошел в УШЭ, а скорее попадают те, кто набрал мало баллов по обоим предметам, поэтому корреляция приходит к 0
